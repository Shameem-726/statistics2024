```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size = 100000 
u = np.random.uniform(size=sample_size) # generating uniform random variables
```

```python
F_inverse = lambda y: np.sqrt(np.log((-1 - y) / (y - 1))) # inverse of CDF
```

```python
sample = F_inverse(u) 
x = np.linspace(0,12,1000)
p = lambda x: (4*x*np.exp(-x**2))/pow((1+np.exp(-x**2)),2) # PDF
```

```python
plt.hist(sample, bins="auto", density=True, label="sample")
plt.plot(x, p(x), linewidth=2, label="PDF") 
plt.legend()
plt.show()
```
