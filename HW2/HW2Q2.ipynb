```python
import numpy as np
import matplotlib.pyplot as plt
```

First part of question


```python
# Sample size which was mentioned
n = 10**5
```

```python
# Draw a sample from N(0, 1)
X = np.random.normal(0, 1, n)
```

```python
# Apply the function g(x) = atan(x)
Y = np.arctan(X)
```

```python
# Plot the histogram of Y
plt.figure(figsize=(10, 6))
plt.hist(Y, bins=100, alpha=0.7, color='yellow', edgecolor='black')
plt.title('Histogram of Y = arctan(X) where X ~ N(0, 1)')
plt.xlabel('Y')
plt.ylabel('Density')
plt.grid()
plt.show()
```

Second part of question

```python
def fun(y):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-np.tan(y)**2 / 2) * (1 / np.cos(y)**2)
```

```python
y_values = np.linspace(-np.pi/2 + 0.01, np.pi/2 - 0.01, 1000)
pdf_values = fun(y_values)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(Y, bins=100, density=True, alpha=0.5, color='yellow', edgecolor='black', label='Histogram of Y')
plt.plot(y_values, pdf_values, color='red', label='PDF of Y', linewidth=2)
plt.title('Histogram and PDF of Y = arctan(X) where X ~ N(0, 1)')
plt.xlabel('Y')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
